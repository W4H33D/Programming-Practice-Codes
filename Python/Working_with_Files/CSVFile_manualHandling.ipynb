### Reading a CSV File Line by Line (Raw Output)

```python
with open('CSVs/sales_data.csv', 'r', encoding='utf-8') as file:
    for line in file:
        print(line)
```

### Inspecting Raw Line Strings with repr()

```python
with open('CSVs/sales_data.csv', 'r', encoding='utf-8') as file:
    for line in file:
        print(repr(line))
```

### Stripping Whitespace and Newline Characters

```python
with open('CSVs/sales_data.csv', 'r', encoding='utf-8') as file:
    for line in file:
        raw_data = line.strip()   # Removes New Line character
        print(raw_data)
```

### Splitting Lines into Fields with .split(',')

```python
with open('CSVs/sales_data.csv', 'r', encoding='utf-8') as file:
    for line in file:
        raw_data = line.strip()   # Removes New Line character
        cleaned_data = raw_data.split(',')
        print(cleaned_data)
```

### Chaining .strip().split(',') in One Step

```python
with open('CSVs/sales_data.csv', 'r', encoding='utf-8') as file:
    for line in file:
        cleaned_data = line.strip().split(',')
        print(cleaned_data)
```

### Chaining With f-String To Showcase Usefull Summaries

```python
with open('CSVs/sales_data.csv', 'r', encoding='utf-8') as file:
    next(file)
    for line in file:
        product, price, quantity, discount = line.strip().split(',')
        print(f"{product} is available in ${int(price):,.2f} | Stocks Left : {int(quantity)} | Discount : {float(discount):.1%}")
            
```
