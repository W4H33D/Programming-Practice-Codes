### Opening And Read File

```python
# open("feedback_students.txt", 'r', 'utf-8')       #Syntax: open(file_path, mode, encoding)
with open('CSVs/feedback_students.txt', 'r', encoding='utf-8') as file:   # safe way to open a file
    # Working With File Code here
    lines = file.readlines()        # readlines() method read every line of the file.
    for l in lines:                 # We loop on each line one by one
        print(l.strip())            # .strip() method removes spaces, tabs, and newline characters from beginning and end
                                    # Note readlines() method good for small file however it doesn't fit for
                                    # Large files because it will take too much memory. The best way to read is:
```

### Recommend Way To Read Files

```python
print("\n== Best Way To Read Files ==\n")
with open('CSVs/feedback_students.txt', 'r', encoding='utf-8') as file:
    for line in file:
        print(line.strip())
```

### Error Handling For File Opening

```python
import os

file_path = "feedback.txt"              # Unknown File Handling Method 1.
if os.path.exists(file_path):           # Checks If the file exits or not
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            print(line.strip())
else:
    print("File does not exist. Please check the filename or path.")
    exit(1)
```

### Recommended Error Handling For File Operations

```python
print("\n== Best Way To Read Files With Error Handling ==\n")
try:
    with open('CSVs/feedback_students.txt', 'r', encoding='utf-8') as file:
        for line in file:
            print(line.strip())
except FileNotFoundError:
    print(f"File (feedback_students.txt) Doesn't Exits. Check For Name or Path")
except Exception as e:
    print(f"Error Occurred: {e}")
else:
    # Execute When Try Block Run Without Any Error
    print("Log   :   File Successfully Read")
finally:
    # Runs For Sure No Matter What Happens
    print("Log   :   Try BLock Is Ended")
```
