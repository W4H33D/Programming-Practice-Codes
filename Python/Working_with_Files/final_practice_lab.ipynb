### Project: The Sales Records Cleanup
**Scenario**
You are working as a Junior Data Analyst for a small online shop. This time, the sales team has exported the raw records into a CSV file instead of giving you a Python list.

The problem is that the file is still messy. Some rows contain extra spaces, some customer names are missing, one record has a missing sales amount, one record has a negative value, and some rows are duplicated.

Your job is to read the raw data from the CSV file, clean it, and then save the final cleaned records into a new CSV file that is ready for analysis.

The raw file is named `lab_raw_sales.csv`, and your cleaned output file should be named `lab_cleaned_sales.csv`.

```python
import csv, os
from datetime import datetime

file_path = 'CSVs/lab_raw_sales.csv'
raw_data = []
clean_data = []
header = []
if os.path.exists(file_path):           # Checks If the file exits or not
    print("File exists. Ready to load.")
else:
    print("File does not exist. Please check the path.")
    exit(1)
```

```python
with open(file_path, 'r', newline="") as csv_file:
    file_content = csv.reader(csv_file)
    header = next(file_content)
    for sale_id, customer, amount, date in file_content:
        raw_data.append({"sale_id": sale_id, "customer": customer, "amount": amount, "date": date})
```

```python
clean_data = [
    {key.strip(): value.strip() for key, value in list_items.items()}    # Removing White Spaces
    for list_items in raw_data
]
```

```python
clean_data = [
    {key: value for key, value in list_items.items()}    # Removing White Spaces
    for list_items in clean_data if list_items['customer'] != '' and list_items['amount'] != "" and float(list_items['amount']) >= 0.0
]

for item in clean_data:
    item['amount'] = float(item['amount']) # Changing Amount value into actual number
    item['sale_id'] = int(item['sale_id'])
```

```python
seen = set()
unique_data = []

for items in clean_data:
    row_tuple = tuple(items.items())
    if row_tuple not in seen:
        seen.add(row_tuple)
        unique_data.append(items)
clean_data = unique_data
del seen, unique_data
```

```python
header = [
    item.title() for item in header
]
clean_data = [
    {key.title(): value for key, value in i.items()}
    for i in clean_data # CLeaned List: Standerized Key values
]
```

```python
for item in clean_data:
    format_code = '%d/%b/%Y'
    parsed_time = datetime.strptime(item['Date'], format_code)
    item['Date'] = parsed_time.strftime('%Y-%d-%m') # CLeaned List Output: Parsed Date
```

```python
for i, data in enumerate(clean_data, start=1):
    print(f"{i:>2}:  {data}")
```

```python
try:
    with open('CSVs/lab_cleaned_sales.csv', 'w', newline="", encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=header)
        writer.writeheader()
        writer.writerows(clean_data)
    print("File saved successfully as sales_report.csv!")
except Exception as e:
    print(f"Error Occured: {e}")
```
